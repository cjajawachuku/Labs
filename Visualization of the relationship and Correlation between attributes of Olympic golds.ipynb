# Visualization of the relationship and Correlation between attributes of Olympic golds

```python
# Load necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('olympics-3.csv')
```

### Exploratory Data Analysis

```python
# View first five rows of the dataset
df.head()
```

### Feature Engineering

```python
# encoding categorical variables as numerical values
df['City'], _ = pd.factorize(df['City'])
df['Sport'], _ = pd.factorize(df['Sport'])
df['Discipline'], _ = pd.factorize(df['Discipline'])
df['Athlete'], _ = pd.factorize(df['Athlete'])
df['NOC'], _ = pd.factorize(df['NOC'])
df['Gender'], _ = pd.factorize(df['Gender'])
df['Event'], _ = pd.factorize(df['Event'])
df['Event_gender'], _ = pd.factorize(df['Event_gender'])
df['Medal'], _ = pd.factorize(df['Medal'])

# printing the dataframe with numerical values
df.sample(10)
```

### Data visualization

```python
# create a correlation matrix to see the correlation between each attribute
plt.figure(figsize=(15,10))
corr = df.corr()
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# visualize the relationship and correlation between all attributes of the Olympic golds dataset
sns.pairplot(df)
plt.show()
```
