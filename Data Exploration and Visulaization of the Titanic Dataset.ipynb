# Data Exploration and Visulaization of the Titanic Dataset

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
titanic_df = pd.read_csv('titanic.csv')
```

### Exploratory Data Analysis

```python
# View first few rows of the dataset
titanic_df.head()
```

```python
# View information about the dataset
titanic_df.info()
```

```python
# View summary statistics of the dataset
titanic_df.describe()
```

```python
# View the number of missing values in each column
titanic_df.isnull().sum()
```

### Data Visualization

```python
# Bar chart of passengers' survival by sex
sns.barplot(x='Sex', y='Survived', data=titanic_df)
plt.show()
```

```python
# Bar chart of passengers' survival by embarked port
sns.barplot(x='Embarked', y='Survived', data=titanic_df)
plt.show()
```

```python
# Boxplot of passengers' ages by survival
sns.boxplot(x='Survived', y='Age', data=titanic_df)
plt.show()
```

```python
# Scatterplot of passengers' ages and fares by survival
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=titanic_df)
plt.show()
```

```python
# Pairplot of passengers' ages, fares, and survival by class
sns.pairplot(data=titanic_df[['Age', 'Fare', 'Survived', 'Pclass']], hue='Pclass')
plt.show()
```
